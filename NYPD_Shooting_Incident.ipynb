```python
# Here we are importing all the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# we use read_csv command to read the data set and store it in the variable called as df
df = pd.read_csv('/content/NYPD_Shooting_Incident_Data__Historic__20240308.csv')
# will give us the intel of the dataset
df
```

```python
#This command will provide information about the dataset such as data type, null count number of column and more.
df.info()
```

```python
# This will provide us with all the unqire values from the given column
df.nunique()
```

```python
#In this process we are dropping all the non required columns as this si the first step of cleansing the program
column_drop = ['LOC_OF_OCCUR_DESC', 'JURISDICTION_CODE', 'LOCATION_DESC']
df.drop(column_drop,axis=1,inplace=True)
df.head()
```

```python
#Here we are calculating the number of null values per column
df.isnull().sum()
```

```python
# Dropping null values from the data set
df = df.dropna()
```

```python
# Finally we got the dataset which has been cleansed
df.isnull().sum()
```

```python
#describe() function gives you information about min, max and mean of all the values in the column
#Used to find outliners but this dataset consist of minimum to no outliners
df.describe()
```

```python
# Here we are removing the duplicate items
df = df.drop_duplicates()
df.head()
```

```python
# Here we are changing our date type from object to datetime64 format where we can get access to month,day and year, In a precise way.
df['OCCUR_DATE'] = df['OCCUR_DATE'].astype(np.datetime64)
df.head()
```

```python
# It defines the shape of the dataset i.e number of rows and column
df.shape
```

```python
# Using seaborn we are using count plot to specify data which represents the value from LOC_CLASSFCTN_DESC i.e classified location according
# To the boro in wich it occured
sns.countplot(x = "LOC_CLASSFCTN_DESC", hue = "BORO", data=df, palette='mako' )
# Changing the figure size
sns.set_theme(rc={'figure.figsize':(20,6)})
```

```python
# Finding all the unique values from PERP_RACE column
df['PERP_RACE'].unique()
```

```python
# Setting the size of the figure
plt.subplots(figsize=(11, 5))
# Creating a new variable which will have a dataframe, this is a dictionary combination that iterates over the group
# For every group created, it calculates the counts of unique values in the 'BORO' column using value_counts()
# And then it stores these counts in a dictionary with the group label 'PERP_RACE'  as the key (x_label).
df_heat = pd.DataFrame({
    x_label: grp['BORO'].value_counts()
    for x_label, grp in df.groupby('PERP_RACE')
})
# Here we are plotting the heat map of multidimeansional data base which we have created
sns.heatmap(df_heat, cmap='crest')
# We are labeling x and y axis
plt.xlabel('ETHNICITY')
plt.ylabel('LOCATION')

```

```python
# Just for the refference purpose we call the first 5 row of the data set again
df.head()
```

```python
# This will create a column in the data base called as "month", And will extract only the months from "OCCUR_DATE" column
df['Month'] = df['OCCUR_DATE'].dt.month

# Count the number of arrests for each month
incident_month = df['Month'].value_counts().sort_index()
print(incident_month)

```

```python
# Here we are using .plt method to plot the graph first mentioning index of the data set then the actual "incident_month dataset"
# Then we add color, linestyle and width of the line, for customization purpose
plt.plot(incident_month.index, incident_month, marker='.', color='red', linestyle='--', linewidth=1)
# Giving title to our graph
plt.title('INCIDENT PER MONTH')
# Here we are just creating a range from (1,13) i.e it will take number from 1-12.
# Then we are assigning numbers to belowe specified string i.e 1='jan', 2='feb'.
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
# Here we are labeling x and y axis
plt.xlabel('MONTH')
plt.ylabel('NUMBER OF INCIDENTS')
# Again changing the figure size
plt.figure(figsize=(12, 6))
# Then finally plotting it
# This data gives the number incident happened per month
plt.show()
```

```python
# Again for reference purpose
df.head()
```

```python
# We have to import a new library used for leaflet map plotting
import folium

# finding mean to create a base map
inc_map = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=12)

# Adding markers for each shooting spot
for a,r in df.iterrows():
    folium.Marker(
        location=[r['Latitude'], r['Longitude']],
        popup=f"{r['LOC_CLASSFCTN_DESC']}\n{r['PERP_SEX']}",
        icon=folium.Icon(color='blue', icon='gun', prefix='fa')
    ).add_to(inc_map)

# saving file as HTML
inc_map.save('shooting_incident_map.html')

# Use to display the map
inc_map
```
